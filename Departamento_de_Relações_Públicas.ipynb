```python
reviews_df = reviews_df.drop(['date', 'rating', 'lenght'], axis=1)
reviews_df
```

```python
from sklearn.preprocessing import OneHotEncoder
onehotencoder = OneHotEncoder()
X_cat = onehotencoder.fit_transform(X_cat).toarray()
```

```python
type(X_cat)
```

```python
X_cat = pd.DataFrame(X_cat)
type(X_cat)
```
